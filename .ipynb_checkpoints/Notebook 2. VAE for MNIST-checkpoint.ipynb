# 2 VAE for MNIST

```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from scipy.stats import norm
```

```python
# import mnist dataset
from keras.datasets import mnist

(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = mnist.load_data()

x_train_mnist = x_train_mnist.astype('float32') / 255.
x_test_mnist  = x_test_mnist.astype('float32') / 255.
x_train_mnist = np.reshape(x_train_mnist, (len(x_train_mnist), 28, 28, 1))
x_test_mnist  = np.reshape(x_test_mnist,  (len(x_test_mnist),  28, 28, 1))

print(x_train_mnist.shape)
print(x_test_mnist.shape)
```

### VAE for MNIST dataset

```python
from keras.layers import Input, Dense 
from keras.layers import BatchNormalization, Dropout, Flatten, Reshape, Lambda
from keras.models import Model

from keras.objectives import mean_squared_error
from keras.objectives import binary_crossentropy
from keras.layers.advanced_activations import LeakyReLU
from keras import backend as K
```

```python
batch_size = 500
latent_dim = 2
dropout_rate = 0.3
start_lr = 0.0001

def create_vae():
    models = {}

    # Encoder
    input_img = Input(batch_shape=(batch_size, 28, 28, 1))
    x = Flatten()(input_img)
    x = Dense(256, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(128, activation='relu')(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)

    # Predict parameters of distribution
    # Predict log of variation instead of std
    z_mean = Dense(2)(x)
    z_log_var = Dense(2)(x)

    # Sampling from Q with reparametrization trick
    def sampling(args):
        z_mean, z_log_var = args
        epsilon = K.random_normal(shape=(batch_size, latent_dim), mean=0.0, stddev=1.0)
        #epsilon = np.random.normal(size=(batch_size, latent_dim))
        return z_mean + K.exp(z_log_var / 2) * epsilon
    
    l = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])
    
    models["encoder"]  = Model(input_img, l, 'Encoder') 
    models["z_meaner"] = Model(input_img, z_mean, 'Enc_z_mean')
    models["z_lvarer"] = Model(input_img, z_log_var, 'Enc_z_log_var')

    # Decoder
    z = Input(shape=(latent_dim, ))
    x = Dense(128)(z)
    x = LeakyReLU()(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(256)(x)
    x = LeakyReLU()(x)
    x = BatchNormalization()(x)
    x = Dropout(dropout_rate)(x)
    x = Dense(28*28, activation='sigmoid')(x)
    decoded = Reshape((28, 28, 1))(x)

    models["decoder"] = Model(z, decoded, name='Decoder')
    models["vae"]     = Model(input_img, models["decoder"](models["encoder"](input_img)), name="VAE")
    
    def vae_loss(x, decoded):
        x = K.reshape(x, shape=(batch_size, 28*28))
        decoded = K.reshape(decoded, shape=(batch_size, 28*28))
        xent_loss = 28*28*binary_crossentropy(x, decoded)
        kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
        return (xent_loss + kl_loss)/2/28/28

    return models, vae_loss

models, vae_loss = create_vae()
vae = models["vae"]
```

```python
from keras.optimizers import Adam, RMSprop

vae.compile(optimizer=Adam(start_lr), loss=vae_loss)
vae.summary()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import clear_output
from keras.callbacks import LambdaCallback, ReduceLROnPlateau, TensorBoard

imgs = x_test_mnist[:batch_size]
digit_size = 28

def plot_digits(*args, invert_colors=False):
    args = [x.squeeze() for x in args]
    n = min([x.shape[0] for x in args])
    figure = np.zeros((digit_size * len(args), digit_size * n))

    for i in range(n):
        for j in range(len(args)):
            figure[j * digit_size: (j + 1) * digit_size,
                   i * digit_size: (i + 1) * digit_size] = args[j][i].squeeze()
    if invert_colors:
        figure = 1-figure
    plt.figure(figsize=(2*n, 2*len(args)))
    plt.imshow(figure, cmap='Greys_r')
    plt.grid(False)
    ax = plt.gca()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    plt.show()

# Function that we will start after every epoch 
def on_epoch_end(epoch, logs):
    if epoch%50 == 0:
        #clear_output() # clean output
        # real and decoded digits
        decoded = vae.predict(imgs, batch_size=batch_size)
        plot_digits(imgs[:10], decoded[:10])
        
# Callbacks
pltfig = LambdaCallback(on_epoch_end=on_epoch_end)
tb     = TensorBoard(log_dir='./logs')
```

```python
# train the VAE with callbacks
vae.fit(x_train_mnist, x_train_mnist, shuffle=True, epochs=1000,
        batch_size=batch_size,
        validation_data=(x_test_mnist, x_test_mnist),
        callbacks=[pltfig, tb],
        verbose=1)

# save model
#vae.save("/home/vlbackend/EURECOM/ML_in_the_kitchen/Models/vae_mnist")
```

```python
# display a 2D manifold of the digits
n = 15  # figure with 15x15 digits
digit_size = 28
generator = models["decoder"]

figure = np.zeros((digit_size * n, digit_size * n))

# linearly spaced coordinates on the unit square were transformed through the inverse CDF (ppf) of the Gaussian
# to produce values of the latent variables z, since the prior of the latent space is Gaussian
grid_x = norm.ppf(np.linspace(0.05, 0.95, n))
grid_y = norm.ppf(np.linspace(0.05, 0.95, n))

# display a 2D manifold of the digits
def draw_manifold(generator, show=True):
    for i, yi in enumerate(grid_x):
        for j, xi in enumerate(grid_y):
            z_sample = np.array([[xi, yi]])
            x_decoded = generator.predict(z_sample)
            digit = x_decoded[0].reshape(digit_size, digit_size)
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

    plt.figure(figsize=(10, 10))
    plt.imshow(figure, cmap='Greys_r')
    plt.show()
    return figure

figure = draw_manifold(generator, show=True)
```
