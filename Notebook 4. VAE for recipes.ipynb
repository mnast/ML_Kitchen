## 4.2 Create model of VAE

```python
from keras.layers import Input, Dense 
from keras.layers import BatchNormalization, Dropout, Flatten, Reshape, Lambda
from keras.models import Model

from keras.objectives import binary_crossentropy
from keras import backend as K
```

```python
# dataset #1
batch_size = 50
latent_dim = 3
input_dim = 1215
intermed_dim = 100
#dropout_rate = 0.1
#start_lr = 0.00001
```

```python
# dataset #3
batch_size = 100
latent_dim = 20
input_dim = 500
intermed_dim = 100
start_lr = 0.0001
```

```python
def create_vae():
    models = {}

    # Encoder
    input_rec = Input(batch_shape=(batch_size, input_dim))
    x = Dense(intermed_dim, activation='relu')(input_rec)
    x = BatchNormalization()(x)

    # Predict parameters of distribution
    # Predict log of variation instead of std
    z_mean = Dense(latent_dim)(x)
    z_log_var = Dense(latent_dim)(x)

    # Sampling from Q with reparametrization trick
    def sampling(args):
        z_mean, z_log_var = args
        epsilon = K.random_normal(shape=(batch_size, latent_dim), mean=0.0, stddev=1.0)
        return z_mean + K.exp(z_log_var / 2) * epsilon
    
    l = Lambda(sampling, output_shape=(latent_dim,))([z_mean, z_log_var])
    
    models["encoder"]  = Model(input_rec, l, 'Encoder') 
    models["z_meaner"] = Model(input_rec, z_mean, 'Enc_z_mean')
    models["z_lvarer"] = Model(input_rec, z_log_var, 'Enc_z_log_var')

    # Decoder
    z = Input(shape=(latent_dim, ))
    x = Dense(intermed_dim, activation='relu')(z)
    x = BatchNormalization()(x)
    x = Dense(input_dim, activation='relu')(x)
    decoded = Reshape((input_dim,))(x)

    models["decoder"] = Model(z, decoded, name='Decoder')
    models["vae"]     = Model(input_rec, models["decoder"](models["encoder"](input_rec)), name="VAE")
    
    def vae_loss(x, decoded):
        x = K.reshape(x, shape=(batch_size, input_dim))
        decoded = K.reshape(decoded, shape=(batch_size, input_dim))
        xent_loss = input_dim*binary_crossentropy(x, decoded)
        kl_loss = -0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
        return (xent_loss + kl_loss)/2/input_dim

    return models

models = create_vae()
vae = models["vae"]
```

```python
from keras.optimizers import Adam, RMSprop
vae.compile(optimizer='Adam', loss='mean_squared_error')

vae.summary()
```

## 4.3 Train model

```python
from keras.callbacks import Callback
from keras.callbacks import TensorBoard

lss = []
val_lss = []

class LossHistory(Callback):
    #def on_train_begin(self, logs={}):
        #self.losses = []
        #self.val_losses = []

    def on_epoch_end(self, batch, logs={}):  #on_epoch_end
        #self.losses.append(logs.get('loss'))   
        #self.val_losses.append(logs.get('val_loss'))
        lss.append(logs.get('loss'))
        val_lss.append(logs.get('val_loss'))

#Callbacks
history = LossHistory()  
tensorboard = TensorBoard(log_dir='logs', write_graph=True)
```

### Train model with Cross-validation:
#### for Dataset #1 (2015)

```python
# cross-validation training
# split dataset in train (4500) and test (1000) datasets 
# after each 10 epoch shift the test dataset to the right by 50 recipes

# split and train
N=100
for n in range(N):
    print("NUMBER of EPOCH = EPOCH x", n )
    
    low_bnd = n*50
    up_bnd = n*50+1000
    
    if up_bnd <= 5500:
        x_test = data[low_bnd:up_bnd,:]
        x_train = np.concatenate((data[0:low_bnd,:], data[up_bnd:,:]), axis=0)
    else:
        x_test = np.concatenate((data[:up_bnd-5500,:],data[low_bnd:5500,:]), axis=0)
        x_train = data[up_bnd-5500:low_bnd,:]
    
    x_train = x_train.astype('int') 
    x_test = x_test.astype('int') 
    
    # train model
    vae.fit(x_train, x_train,
        epochs=10,
        batch_size=50,
        shuffle=True,
        validation_data=(x_test, x_test),
        callbacks=[history, tensorboard],
        verbose=1)   
```

```python
import math

# if of nearest recipes
def nearest_recipes(p, pp, num_neighbours, dim):
    # for 3D
    if dim == 3:
        distance = [math.sqrt((p[0]-pp[i][0])**2 + (p[1]-pp[i][1])**2 + (p[2]-pp[i][2])**2) for i in range(len(pp))]
    # for 20D
    elif dim == 20:
        distance = [ math.sqrt((p[0]-pp[i][0])**2 + (p[1]-pp[i][1])**2 + (p[2]-pp[i][2])**2 + \
            (p[3]-pp[i][3])**2 + (p[4]-pp[i][4])**2 + (p[5]-pp[i][5])**2 + (p[6]-pp[i][6])**2 + \
            (p[7]-pp[i][7])**2 + (p[8]-pp[i][8])**2 + (p[9]-pp[i][9])**2 + (p[10]-pp[i][10])**2 + \
            (p[11]-pp[i][11])**2 + (p[12]-pp[i][12])**2 + (p[13]-pp[i][13])**2 + (p[14]-pp[i][14])**2 + \
            (p[15]-pp[i][15])**2 + (p[16]-pp[i][16])**2 + (p[17]-pp[i][17])**2 + (p[18]-pp[i][18])**2 + \
            (p[19]-pp[i][19])**2) for i in range(len(pp)) ]
    else:
        print('cannot compute for this dimention')

    neighbours = sorted(zip(distance, range(1000)), key=lambda x: x[0])
    nearest_neighbours = [ neighbours[1:num_neighbours+1][i][1] for i in range(num_neighbours) ]
    
    return nearest_neighbours
                                                                           
nearest_recs = [] 
num_neighbours = 10

def get_coordinates(points, x_test_encoded, dim):
    # for 3D
    if dim == 3:
        coord = [ [x_test_encoded[points[i],0],x_test_encoded[points[i],1],x_test_encoded[points[i],2]] \
                 for i in range(len(points)) ]
    # for 20D
    elif dim == 20:
        coord = [ [x_test_encoded[points[i],0],x_test_encoded[points[i],1],x_test_encoded[points[i],2],x_test_encoded[points[i],3],\
              x_test_encoded[points[i],4],x_test_encoded[points[i],5],x_test_encoded[points[i],6],x_test_encoded[points[i],7],\
              x_test_encoded[points[i],8],x_test_encoded[points[i],9],x_test_encoded[points[i],10],x_test_encoded[points[i],11],\
              x_test_encoded[points[i],12],x_test_encoded[points[i],13],x_test_encoded[points[i],14],x_test_encoded[points[i],15],\
              x_test_encoded[points[i],16],x_test_encoded[points[i],17],x_test_encoded[points[i],18],x_test_encoded[points[i],19]] \
                 for i in range(len(points)) ]
    else:
        print('cannot compute for this dimention')
    return coord
        
coord = get_coordinates(points, x_test_encoded, dim=20)
for i in range(len(points)):
    nearest_recs.append(nearest_recipes(coord[i], x_test_encoded, num_neighbours=num_neighbours,dim=20))
    
print(nearest_recs)
```
